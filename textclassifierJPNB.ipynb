```python
#import req pkgs
import pandas as pd
import numpy as np
import seaborn as sns
import neattext.functions as nfx
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv("emotion_dataset_2.csv")
```

```python
df.head()
```

```python
df['Emotion'].value_counts()
```

```python
sns.countplot(x='Emotion',data=df)
```

```python
dir(nfx)
```

```python
df['Text'].apply(nfx.remove_userhandles)
```

```python
df['Clean_Text']=df['Text'].apply(nfx.remove_userhandles)
```

```python
df['Clean_Text']=df['Clean_Text'].apply(nfx.remove_stopwords)
```

```python
df
```

```python
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

```python
x_train,x_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.3,random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
#logistic regression pipeline
pipe_lr = Pipeline(steps=[('cv',CountVectorizer()),('lr',LogisticRegression())])
```

```python
#Train and Fit Data
pipe_lr.fit(x_train,y_train)
```

```python
#check Accuracy
pipe_lr.score(x_test,y_test)
```

```python
#Testing program with this application
ex1= "aaron is disgusting guy"
```

```python
pipe_lr.predict([ex1])
```

```python
pipe_lr.predict_proba([ex1])
var1=pipe_lr.predict_proba([ex1])
```

```python
#block to save model and pipeline
import joblib
pipeline_file = open("emotion_classifier_model.pkl","wb")
joblib.dump(pipe_lr,pipeline_file)
pipeline_file.close()
```
